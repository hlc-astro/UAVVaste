```python
import json
import os

# Define complete paths to the input and output JSON files
annotations_json_path = '../annotations/annotations.json'
split_json_path = '../annotations/train_val_test_distribution_file.json'

# Load the train_val_test_distribution_file.json to get the list of train images
with open(split_json_path) as train_val_test_file:
    train_val_test_data = json.load(train_val_test_file)

# Get the list of train images from the train split information
train_images = train_val_test_data.get("train", [])
# Get the list of test images from the test split information
test_images = train_val_test_data.get("test", [])
# Get the list of validation images from the validation split information
val_images = train_val_test_data.get("val", [])

# Load the annotations.json file
with open(annotations_json_path) as annotations_file:
    annotations_data = json.load(annotations_file)

# Filter the annotations based on the test split information
test_annotations = [ann for ann in annotations_data['annotations'] if annotations_data['images'][ann['image_id']]['file_name'] in test_images]
train_annotations = [ann for ann in annotations_data['annotations'] if annotations_data['images'][ann['image_id']]['file_name'] in train_images]
val_annotations = [ann for ann in annotations_data['annotations'] if annotations_data['images'][ann['image_id']]['file_name'] in val_images]

# Function to get file names in the directory
def get_files_in_directory(directory_path):
    return os.listdir(directory_path)

# Get the file names in each directory
test_files = get_files_in_directory('/Volumes/Samsung_USB/test_train_val/images/test')  # Replace with the actual path to the test directory
train_files = get_files_in_directory('/Volumes/Samsung_USB/test_train_val/images/train')  # Replace with the actual path to the train directory
val_files = get_files_in_directory('/Volumes/Samsung_USB/test_train_val/images/val')  # Replace with the actual path to the validation directory

# Filter the test_annotations, train_annotations, and val_annotations based on the files present in their respective directories
test_annotations = [ann for ann in test_annotations if annotations_data['images'][ann['image_id']]['file_name'] in test_files]
train_annotations = [ann for ann in train_annotations if annotations_data['images'][ann['image_id']]['file_name'] in train_files]
val_annotations = [ann for ann in val_annotations if annotations_data['images'][ann['image_id']]['file_name'] in val_files]

# Create the reduced annotations data with test images and test annotations
test_reduced_annotations_data = {
    "images": [annotations_data['images'][ann['image_id']] for ann in test_annotations],
    "categories": annotations_data['categories'],
    "annotations": test_annotations
}

# Save the reduced_annotations_data as a JSON file for the test split
with open('test_reduced_annotations.json', 'w') as reduced_annotations_file:
    json.dump(test_reduced_annotations_data, reduced_annotations_file, indent=2)

# Create the reduced annotations data with train images and test annotations
train_reduced_annotations_data = {
    "images": [annotations_data['images'][ann['image_id']] for ann in train_annotations],
    "categories": annotations_data['categories'],
    "annotations": train_annotations
}

# Save the reduced_annotations_data as a JSON file for the train split
with open('train_reduced_annotations.json', 'w') as reduced_annotations_file:
    json.dump(train_reduced_annotations_data, reduced_annotations_file, indent=2)

# Create the reduced annotations data with val images and test annotations
val_reduced_annotations_data = {
    "images": [annotations_data['images'][ann['image_id']] for ann in val_annotations],
    "categories": annotations_data['categories'],
    "annotations": val_annotations
}

# Save the reduced_annotations_data as a JSON file for the validation split
with open('val_reduced_annotations.json', 'w') as reduced_annotations_file:
    json.dump(val_reduced_annotations_data, reduced_annotations_file, indent=2)
```

```python
import json
import os

# Function to get file names in the directory
def get_files_in_directory(directory_path):
    return os.listdir(directory_path)

# Function to get the number of files in a directory
def get_num_files_in_directory(directory_path):
    return len(get_files_in_directory(directory_path))

# Define complete paths to the directories
test_directory_path = '/Volumes/Samsung_USB/test_train_val/images/test'  # Replace with the actual path to the test directory
train_directory_path = '/Volumes/Samsung_USB/test_train_val/images/train'  # Replace with the actual path to the train directory
val_directory_path = '/Volumes/Samsung_USB/test_train_val/images/val'  # Replace with the actual path to the validation directory

# Load the previously created JSON files
with open('test_reduced_annotations.json') as test_json_file:
    test_data = json.load(test_json_file)

with open('train_reduced_annotations.json') as train_json_file:
    train_data = json.load(train_json_file)

with open('val_reduced_annotations.json') as val_json_file:
    val_data = json.load(val_json_file)

# Get the list of file names in the directories
test_files_in_directory = get_files_in_directory(test_directory_path)
train_files_in_directory = get_files_in_directory(train_directory_path)
val_files_in_directory = get_files_in_directory(val_directory_path)

# Get the list of file names in the JSON files
test_files_in_json = [img_info['file_name'] for img_info in test_data['images']]
train_files_in_json = [img_info['file_name'] for img_info in train_data['images']]
val_files_in_json = [img_info['file_name'] for img_info in val_data['images']]

# Check if the file names in the JSON files match the files present in the directories
if set(test_files_in_directory) == set(test_files_in_json):
    print("File names in test JSON match files in the test directory.")
else:
    print("File names in test JSON do not match files in the test directory.")

if set(train_files_in_directory) == set(train_files_in_json):
    print("File names in train JSON match files in the train directory.")
else:
    print("File names in train JSON do not match files in the train directory.")

if set(val_files_in_directory) == set(val_files_in_json):
    print("File names in validation JSON match files in the validation directory.")
else:
    print("File names in validation JSON do not match files in the validation directory.")
```

```python
# Find the files in the train JSON that are not in the train directory
train_files_not_found = set(train_files_in_json) - set(train_files_in_directory)
print("Files in train JSON but not found in train directory:")
print(train_files_not_found)

# Find the files in the val JSON that are not in the val directory
val_files_not_found = set(val_files_in_json) - set(val_files_in_directory)
print("Files in validation JSON but not found in validation directory:")
print(val_files_not_found)
```
